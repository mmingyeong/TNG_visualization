```python
seondeok_snapshot_file_path = glob.glob(f"/home/users/mmingyeong/tng/tng_99_240425/tng_local/snapshot-99.*.hdf5")

num = 0
f = h5py.File(seondeok_snapshot_file_path[num], 'r')
PartType1 = f['PartType1']
dm_pos = PartType1["Coordinates"][:]
dm_pos = dm_pos*0.001 # kpc -> Mpc 단위 변경
dm_pos = dm_pos.astype(np.float64) # 데이터 타입 변경
```

```python
# Example Scripts - Snapshot
# load the spatial Coordinates of all the dark matter particles in the box, and make a quick image with a 2D histogram, projecting out the z-axis.
fig = plt.figure(figsize=(8,8))
h = plt.hist2d(dm_pos[:, 0], dm_pos[:, 1], norm=mpl.colors.LogNorm(), bins=1000)
# norm=mpl.colors.LogNorm(), bins=64
plt.xlim([0,100])
plt.ylim([0, 200])
plt.xlabel('x [cMpc/h]')
plt.ylabel('y [cMpc/h]')
plt.title(f"file {num} projecting out the z-axis.")
cur_ax = plt.gca()
fig.colorbar(h[3],ax=cur_ax)
```

```python
fig = plt.figure(figsize=(8,8))
h=plt.hist2d(dm_pos[:, 1], dm_pos[:, 2], norm=mpl.colors.LogNorm(), bins=1000)
plt.xlim([0,250])
plt.ylim([0,250])
plt.xlabel('y [cMpc/h]')
plt.ylabel('z [cMpc/h]')
plt.title(f"file {num} projecting out the x-axis.")
cur_ax = plt.gca()
fig.colorbar(h[3],ax=cur_ax)
```
