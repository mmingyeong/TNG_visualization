```python
import cairosvg
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# SVG 파일 열기
svg_file = "/home/users/mmingyeong/TNG_visualization/prac/2d_histogram_rasterized_ver1_bins=10_ver3.svg"

# SVG 파일을 이미지로 변환하여 임시 파일에 저장
temp_image_file = "/home/users/mmingyeong/TNG_visualization/prac/test.png"
cairosvg.svg2png(url=svg_file, write_to=temp_image_file)

# 이미지 파일 열기
img = mpimg.imread(temp_image_file)

# 이미지 표시
plt.imshow(img)
plt.axis('off')  # 축 숨기기
plt.show()
```
