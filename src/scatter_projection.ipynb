```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-
#
# @Author: Mingyeong Yang (mmingyeong@kasi.re.kr)
# @Date: 2024-05-27
# @Filename: hist2d_projection.ipynb

import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import glob
import os

import pandas as pd
import h5py
import illustris_python as il
```

```python
num = 1
seondeok_snapshot_file_path = glob.glob(f"/home/users/mmingyeong/tng/tng_99_240425/tng_local/snapshot-99.{num}.hdf5")
seondeok_snapshot_file=f"/home/users/mmingyeong/tng/tng_99_240425/tng_local/snapshot-99.{num}.hdf5"

f = h5py.File(seondeok_snapshot_file, 'r')
PartType1 = f['PartType1']
dm_pos = PartType1["Coordinates"][:]
dm_pos = dm_pos*0.001 # kpc -> Mpc 단위 변경
dm_pos = dm_pos.astype(np.float64) # 데이터 타입 변경
```

```python
fig = plt.figure(figsize=(8, 8))
sc = plt.scatter(dm_pos[:, 0], dm_pos[:, 1], c=dm_pos[:, 2], s=0.1, cmap='viridis')
plt.xlabel('x [cMpc/h]')
plt.ylabel('y [cMpc/h]')
plt.title(f"file {num} projecting out the z-axis.")
plt.colorbar(sc, ax=plt.gca(), label='z [cMpc/h]')  # Add a color bar with a label for the z-axis

plt.show()
```

```python
fig = plt.figure(figsize=(8, 8))
sc = plt.scatter(dm_pos[:, 1], dm_pos[:, 2], c=dm_pos[:, 0], s=0.1, cmap='viridis')
plt.xlabel('y [cMpc/h]')
plt.ylabel('z [cMpc/h]')
plt.title(f"file {num} projecting out the x-axis.")
plt.colorbar(sc, ax=plt.gca(), label='x [cMpc/h]')  # Add a color bar with a label for the z-axis

plt.show()
```

```python
fig = plt.figure(figsize=(8, 8))
sc = plt.scatter(dm_pos[:, 0], dm_pos[:, 2], c=dm_pos[:, 1], s=0.1, cmap='viridis')
plt.xlabel('x [cMpc/h]')
plt.ylabel('z [cMpc/h]')
plt.title(f"file {num} projecting out the y-axis.")
plt.colorbar(sc, ax=plt.gca(), label='y [cMpc/h]')  # Add a color bar with a label for the z-axis

plt.show()
```
