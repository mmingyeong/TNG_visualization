```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-
#
# @Author: Mingyeong Yang (mmingyeong@kasi.re.kr)
# @Date: 2024-05-22
# @Filename: 240522_PlotAllpoint.ipynb

import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import glob
import os

import pandas as pd
import h5py
import illustris_python as il
```

```python
seondeok_snapshot_file_path = glob.glob(f"../../tng/tng_99_240425/tng_local/snapshot-99.*.hdf5")
```

```python
f = h5py.File(seondeok_snapshot_file_path[0], 'r')
PartType1 = f['PartType1']
dm_pos = PartType1["Coordinates"][:]
dm_pos = dm_pos*0.001 # kpc -> Mpc 단위 변경
dm_pos = dm_pos.astype(np.float64) # 데이터 타입 변경
```

```python
dm_pos
```

```python
# Example Scripts - Snapshot
# load the spatial Coordinates of all the dark matter particles in the box, and make a quick image with a 2D histogram, projecting out the z-axis.
fig = plt.figure(figsize=(8,8))
h = plt.hist2d(dm_pos[:, 0], dm_pos[:, 1], norm=mpl.colors.LogNorm(), bins=50)
# norm=mpl.colors.LogNorm(), bins=64
plt.xlim([0,100])
plt.ylim([0, 200])
plt.xlabel('x [cMpc/h]')
plt.ylabel('y [cMpc/h]')
plt.title("projecting out the z-axis.")
cur_ax = plt.gca()
fig.colorbar(h[3],ax=cur_ax)
```

```python
fig = plt.figure(figsize=(8,8))
h=plt.hist2d(dm_pos[:, 1], dm_pos[:, 2], norm=mpl.colors.LogNorm(), bins=50)
plt.xlim([0,250])
plt.ylim([0,250])
plt.xlabel('y [cMpc/h]')
plt.ylabel('z [cMpc/h]')
plt.title("projecting out the x-axis.")
cur_ax = plt.gca()
fig.colorbar(h[3],ax=cur_ax)
```

```python
fig = plt.figure(figsize=(8,8))
h=plt.hist2d(dm_pos[:, 0], dm_pos[:, 2], norm=mpl.colors.LogNorm(), bins=64)
plt.xlim([0,100])
plt.ylim([0, 200])
plt.xlabel('x [cMpc/h]')
plt.ylabel('z [cMpc/h]')
plt.title("projecting out the y-axis.")
cur_ax = plt.gca()
fig.colorbar(h[3],ax=cur_ax)
```

```python
# 3D 그래프를 위한 figure와 3D 축 생성
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 3D 산점도 그리기
scatter = ax.scatter(dm_pos[:, 0], dm_pos[:, 1], dm_pos[:, 2], s=0.01, alpha=0.6, cmap='plasma')

# 축 레이블 설정
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

# 그래프 출력
plt.show()
```
